```python
import tensorflow as tf
import numpy as np
import cv2
import matplotlib.pyplot as plt
from face_alignment import align_face
from landmarks_detector import LandmarksDetector
```

```python
image_path = "/content/-2147483648_-210099.jpg"
```

```python
width = height = 100
```

```python
landmarks_detector = LandmarksDetector('/content/drive/MyDrive/Saved_Models/shape_predictor_68_face_landmarks.dat')
```

```python
face_landmarks = landmarks_detector.get_landmarks(image_path)
```

```python
for face_landmark in face_landmarks:
    image = align_face(image_path, face_landmark)
```

```python
image_show = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.resize(image, (width, height))
image = image / 255.0
image = image.reshape(1, width, height, 3)  #چون تصویر ورودی شبکه به صورت بچ و چهار بعدی بوده در اینجا هم باید تصویر را چهاربعدی به شبکه بدهیم
```

```python
model = tf.keras.models.load_model("/content/drive/MyDrive/Saved_Models/age_prediction_2.h5")
```

```python
output = model.predict(image)
print(output)

img_show = cv2.cvtColor(image_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show)
```
